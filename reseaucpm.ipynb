```python
def autopct(pct): # only show the label when it's > 5%
    return ('%.2f%%' % pct) if pct > 5 else ''

colors = ['cadetblue', 'teal', 'darkcyan', 'mediumturquoise', 'skyblue', 'lightblue',
          'palegreen', 'lightgreen', 'darkseagreen', 'seagreen', 'forestgreen', 'darkgreen']
    
my_labels = ('Grand Casablanca-Settat', 'Rabat-Salé-Kénitra', 'Fès-Meknès', 'Oriental', 'Tanger-Tétouan-Al Hoceima', 'Marrakech-Safi', 
             'Souss-Massa', 'Béni Mellal-Khénifra', 'Drâa-Tafilalet', 'Laayoune-Sakia El Hamra', 'Guelmim-Oued Noun', 'Eddakhla-Oued Eddahab')

reseaucpm['REGION_ADMINISTRATIVE'].value_counts().plot(kind='pie', figsize=(10,12), autopct=autopct, labels=None, colors=colors)
plt.ylabel('')
plt.legend(loc=2, labels=my_labels)
```
